# Context
This is the dataset of 13th edition of the IPL  played in UAE (United Arab Emirates) IPL 2020 had been scheduled between 19 September 2020 and 10 November 2020.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
df['Dream11_ Points'].unique()
```

```python
df[df['Dream11_ Points'].isnull()]
```

we don't have any record for row no 1320 so we can delete this row 

```python
df.iloc[1320:]
```

```python
#droping row no 1320 from dataset 
df=df.drop(1320,axis=0)
df.head()
```

```python
df['Wickets'].unique()
```

# Replacing null values with zero

```python
df=df.fillna(0)
df.isnull().sum()
```

```python
df.info()
```

wickets column are in object first conver in to float 

# Remove unwanted character from data

```python
df['Wickets']=df['Wickets'].str.replace(' ','')


```

```python
df['Wickets']=df['Wickets'].replace('',np.nan)
df['Wickets'].unique()
```

```python
df['Wickets']=df['Wickets'].fillna(0)
df['Wickets']=df['Wickets'].astype('float')
```

# Checking Teamwise

1.which player has scored more run nd more points

2.which player has taken more wickets nd more point

3.which player are continously hitting run for chennai

4.which player are continously taking wickets for chennai


```python
csk_df=df[df['Team']=='CSK']
csk_df
```

```python
plr_run=csk_df.groupby(['Player'])['Runs'].sum().sort_values(ascending=False)
plr_run
```

```python
plr_avg_runs=csk_df.groupby(['Player'])['Runs'].mean().sort_values(ascending=False)
plr_avg_runs
```

```python
plr_wicket=csk_df.groupby(['Player'])['Wickets'].sum().sort_values(ascending=False)
plr_wicket
```

```python
avg_wicket=csk_df.groupby(['Player'])['Wickets'].mean().sort_values(ascending=False)
avg_wicket
```

```python
plr_point=csk_df.groupby(['Player'])['Dream11_ Points'].sum().sort_values(ascending=False)
plr_point
```

```python
plr_avg_point=csk_df.groupby(['Player'])['Dream11_ Points'].mean().sort_values(ascending=False)
plr_avg_point
```

```python
plr_role=csk_df.groupby(['Player'])['Role'].unique()
plr_role
```

# Add all individual series and creat a single data frame of all player

```python
csk_plr_df=pd.concat([plr_run,plr_avg_runs,plr_wicket,avg_wicket,plr_point,plr_avg_point,plr_role],axis=1)
csk_plr_df
```

In this dataframe most of the columns are repeating lets fix this 

```python
csk_plr_df['Total_Runs']=csk_plr_df.iloc[:,0]
csk_plr_df['Avg_Run']=csk_plr_df.iloc[:,1]
csk_plr_df=csk_plr_df.drop('Runs',axis=1)
csk_plr_df['Avg_Wicket']=csk_plr_df.iloc[:,1]
csk_plr_df['Total_Wickets']=csk_plr_df.iloc[:,0]
csk_plr_df=csk_plr_df.drop('Wickets',axis=1)
csk_plr_df['Avg_Point']=csk_plr_df.iloc[:,1]
csk_plr_df['Total_Point']=csk_plr_df.iloc[:,0]
csk_plr_df=csk_plr_df.drop('Dream11_ Points',axis=1)
csk_plr_df['Player_Role']=csk_plr_df.iloc[:,0]
csk_plr_df=csk_plr_df.drop('Role',axis=1)
csk_plr_df

```
